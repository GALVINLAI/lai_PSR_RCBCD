# Lemma
Gavin arbitrary $\lambda, \Omega \in \mathbb{R}$, we let $f_{\cos }(x):=\lambda \cos (\Omega x)$ and $f_{\sin }(x):=\lambda \cos (\Omega x)$. Then we have
$$
\begin{aligned}
\frac{d}{d x} f_{\cos }(x) & =\frac{\Omega}{2 \sin (\Omega s)}\left[f_{\cos }(x+s)-f_{\cos }(x-s)\right] \\
\frac{d}{d x} f_{\sin }(x) & =\frac{\Omega}{2 \sin (\Omega s)}\left[f_{\sin }(x+s)-f_{\sin }(x-s)\right]
\end{aligned}
$$
hold for all $x \in \mathbb{R}$ and $s \neq \frac{\pi}{\Omega} k, k \in \mathbb{Z}$.

# Proof
Gavin arbitrary $\lambda, \Omega \in \mathbb{R}$, we let $f_{\cos }(x):=\lambda \cos (\Omega x)$ and $f_{\sin }(x):=\lambda \cos (\Omega x)$. Then we have
$$
\frac{d}{d x} f_{\cos }(x)=-\lambda \Omega \sin (\Omega x), \quad \frac{d}{d x} f_{\sin }(x)=\lambda \Omega \cos (\Omega x)
$$

We will use the trigonometric identities
$$
\begin{aligned}
\cos (\Omega x+\Omega s) & =\cos (\Omega x) \cos (\Omega s)-\sin (\Omega x) \sin (\Omega s) \\
\cos (\Omega x-\Omega s) & =\cos (\Omega x) \cos (\Omega s)+\sin (\Omega x) \sin (\Omega s) \\
\sin (\Omega x+\Omega s) & =\sin (\Omega x) \cos (\Omega s)+\cos (\Omega x) \sin (\Omega s) \\
\sin (\Omega x-\Omega s) & =\sin (\Omega x) \cos (\Omega s)-\cos (\Omega x) \sin (\Omega s)
\end{aligned}
$$

Then,
$$
\frac{(1)-(2)}{2 \sin (\Omega s)}=\frac{-2 \sin (\Omega x) \sin (\Omega s)}{2 \sin (\Omega s)}=-\sin (\Omega x)=\frac{1}{\lambda \Omega} \frac{d}{d x} f_{\cos }(x),
$$
hence, we claim that
$$
\begin{aligned}
\frac{d}{d x} f_{\cos }(x) & =\frac{\Omega}{2 \sin (\Omega s)}[\lambda \cos (\Omega(x+s))-\lambda \cos (\Omega(x-s))] \\
& =\frac{\Omega}{2 \sin (\Omega s)}\left[f_{\cos }(x+s)-f_{\cos }(x-s)\right]
\end{aligned}
$$

Similarly,
$$
\frac{(3)-(4)}{2 \sin (\Omega s)}=\frac{2 \cos (\Omega x) \sin (\Omega s)}{2 \sin (\Omega s)}=\cos (\Omega x)=\frac{1}{\lambda \Omega} \frac{d}{d x} f_{\sin }(x)
$$
hence, we claim that
$$
\begin{aligned}
\frac{d}{d x} f_{\sin }(x) & =\frac{\Omega}{2 \sin (\Omega s)}[\lambda \sin (\Omega(x+s))-\lambda \sin (\Omega(x-s))] \\
& =\frac{\Omega}{2 \sin (\Omega s)}\left[f_{\sin }(x+s)-f_{\sin }(x-s)\right]
\end{aligned}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter settings
lambda_val = np.random.rand()
Omega = 2.0
s = 1.0
x = np.linspace(0, 10, 1000)

# Function definitions
def f_cos(x):
    return lambda_val * np.cos(Omega * x)

def f_sin(x):
    return lambda_val * np.sin(Omega * x)

# Numerical derivative calculation
def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

# Verification of the proposition expression
def psr_cos_derivative(x):
    return Omega / (2 * np.sin(Omega * s)) * (f_cos(x + s) - f_cos(x - s))

def psr_sin_derivative(x):
    return Omega / (2 * np.sin(Omega * s)) * (f_sin(x + s) - f_sin(x - s))

# Calculate numerical derivatives and expression results
num_deriv_cos = numerical_derivative(f_cos, x)
expr_cos = psr_cos_derivative(x)

num_deriv_sin = numerical_derivative(f_sin, x)
expr_sin = psr_sin_derivative(x)

# Plot the results
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(x, num_deriv_cos, label="Numerical Derivative of $f_{\cos}(x)$", linestyle='dashed')
plt.plot(x, expr_cos, label="PSR Expression of $f_{\cos}'(x)$", linestyle='dotted')
plt.xlabel('$x$')
plt.ylabel('Derivative')
plt.title('Verification of $f_{\cos}(x)$')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, num_deriv_sin, label="Numerical Derivative of $f_{\sin}(x)$", linestyle='dashed')
plt.plot(x, expr_sin, label="PSR xpression of $f_{\sin}'(x)$", linestyle='dotted')
plt.xlabel('$x$')
plt.ylabel('Derivative')
plt.title('Verification of $f_{\sin}(x)$')
plt.legend()

plt.tight_layout()
plt.show()
```
